```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = cv2.imread("lena.jpg", 1)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
imgInfo = imgRGB.shape
height = imgInfo[0]
width = imgInfo[1]
mode = imgInfo[2] 
```

# 图像平移

```python
result = np.zeros(imgInfo, np.uint8)
for i in range(height):
    for j in range(width - 50):
        result[i, j + 50] = imgRGB[i, j]
plt.figure(figsize=(8, 12))
plt.subplot(1,2,1), plt.imshow(imgRGB), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(result), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

# 图像镜像

```python
size = (width,height)
iLR = np.zeros([height, width, mode], np.uint8)
for i in range(height):
    for j in range(width):
        iLR[i,width-1-j] = imgRGB[i,j]
          
plt.figure(figsize=(8, 12))
plt.subplot(1,2,1), plt.imshow(imgRGB), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(iLR), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

# 图像旋转

```python
angle=30
anglePi = angle * math.pi / 180.0
cosA = math.cos(anglePi)
sinA = math.sin(anglePi)
X1 = math.ceil(abs(0.5 * height * cosA + 0.5 * width * sinA))
X2 = math.ceil(abs(0.5 * height * cosA - 0.5 * width * sinA))
Y1 = math.ceil(abs(-0.5 * height * sinA + 0.5 * width * cosA))
Y2 = math.ceil(abs(-0.5 * height * sinA - 0.5 * width * cosA))
hh = int(2 * max(Y1, Y2))
ww = int(2 * max(X1, X2))
result = np.zeros([hh, ww, mode], np.uint8)
for i in range(hh):
    for j in range(ww):
        x = cosA * i + sinA * j - 0.5 * ww * cosA - 0.5 * hh * sinA + 0.5 * width
        y = cosA * j- sinA * i+ 0.5 * ww * sinA - 0.5 * hh * cosA + 0.5 * height
        x = int(x)
        y = int(y)
        if x > -1 and x < height and y > -1 and y < width :
            result[i, j] = imgRGB[x, y]
plt.figure(figsize=(8, 12))
plt.subplot(1,2,1), plt.imshow(imgRGB), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(result), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

# 复合变换

```python
move = np.zeros(imgInfo, np.uint8)
for i in range(height):
    for j in range(width - 10):
        move[i, j + 10] = imgRGB[i, j]
        
size = (width,height)
iLR = np.zeros([height, width, mode], np.uint8)
for i in range(height):
    for j in range(width):
        iLR[i,width-1-j] = move[i,j]
        
angle=30
anglePi = angle * math.pi / 180.0
cosA = math.cos(anglePi)
sinA = math.sin(anglePi)
X1 = math.ceil(abs(0.5 * height * cosA + 0.5 * width * sinA))
X2 = math.ceil(abs(0.5 * height * cosA - 0.5 * width * sinA))
Y1 = math.ceil(abs(-0.5 * height * sinA + 0.5 * width * cosA))
Y2 = math.ceil(abs(-0.5 * height * sinA - 0.5 * width * cosA))
hh = int(2 * max(Y1, Y2))
ww = int(2 * max(X1, X2))
result = np.zeros([hh, ww, mode], np.uint8)
for i in range(hh):
    for j in range(ww):
        x = cosA * i + sinA * j - 0.5 * ww * cosA - 0.5 * hh * sinA + 0.5 * width
        y = cosA * j- sinA * i+ 0.5 * ww * sinA - 0.5 * hh * cosA + 0.5 * height
        x = int(x)
        y = int(y)
        if x > -1 and x < height and y > -1 and y < width :
            result[i, j] = iLR[x, y]
plt.figure(figsize=(8, 12))
plt.subplot(1,2,1), plt.imshow(imgRGB), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(result), plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```
